```python
import random

import numpy as np

import matplotlib.pyplot as plt

import pandas as pd



# Configurações iniciais

tamanho_população = 100

número_de_gerações = 50

tamanho_seleção = 20

taxa_mutação = 0.1



# Representação dos candidatos (id, partido, votos, qualidade)

# Para simulação com 513 deputados, precisamos de muitos mais candidatos.

candidatos = [{'id': f'{chr(65 + i//26)}{i % 26 + 1}', 'partido': chr(65 + i//26), 'votos': random.randint(1000, 10000), 'qualidade': random.uniform(0.5, 1.0)}

              for i in range(2000)]  # 2000 candidatos para garantir variação



número_de_cadeiras = 513

votos_totais = sum(c['votos'] for c in candidatos)

quociente_eleitoral = votos_totais / número_de_cadeiras



def inicializar_população(tamanho):

    return [random.sample(candidatos, número_de_cadeiras) for _ in range(tamanho)]



def calcular_aptidão(indivíduo):

    votos_totais = sum(c['votos'] for c in indivíduo)

    qualidade_total = sum(c['qualidade'] for c in indivíduo)

    return votos_totais + qualidade_total * 10000  # Ajuste o peso da qualidade conforme necessário



def seleção(população, aptidão_população):

    selecionados = random.choices(população, weights=aptidão_população, k=tamanho_seleção)

    return selecionados



def cruzamento(pais):

    filhos = []

    for i in range(len(pais) // 2):

        pai1 = pais[i]

        pai2 = pais[len(pais) - i - 1]

        ponto_corte = random.randint(1, número_de_cadeiras - 1)

        filho = pai1[:ponto_corte] + pai2[ponto_corte:]

        filhos.append(filho)

    return filhos



def mutação(filhos):

    for filho in filhos:

        if random.random() < taxa_mutação:

            indice = random.randint(0, número_de_cadeiras - 1)

            novo_candidato = random.choice(candidatos)

            while novo_candidato in filho:

                novo_candidato = random.choice(candidatos)

            filho[indice] = novo_candidato

    return filhos



def elitismo(população, aptidão_população, mutantes, nova_aptidão):

    nova_população = mutantes + população

    nova_aptidão_total = nova_aptidão + aptidão_população

    melhores_indices = np.argsort(nova_aptidão_total)[-tamanho_população:]

    nova_população = [nova_população[i] for i in melhores_indices]

    nova_aptidão_total = [nova_aptidão_total[i] for i in melhores_indices]

    return nova_população, nova_aptidão_total



def encontrar_melhor_indivíduo(população, aptidão_população):

    melhor_indice = np.argmax(aptidão_população)

    return população[melhor_indice], aptidão_população[melhor_indice]



# Inicializar população e calcular aptidão inicial

população = inicializar_população(tamanho_população)

aptidão_população = [calcular_aptidão(ind) for ind in população]



# Armazenar histórico para plotagem

histórico_aptidão = []



for geração in range(número_de_gerações):

    # Seleção

    pais = seleção(população, aptidão_população)

    # Cruzamento

    filhos = cruzamento(pais)

    # Mutação

    mutantes = mutação(filhos)

    # Avaliar nova população

    nova_aptidão = [calcular_aptidão(ind) for ind in mutantes]

    # Elitismo

    população, aptidão_população = elitismo(população, aptidão_população, mutantes, nova_aptidão)

    # Registrar aptidão média da geração

    aptidão_média = np.mean(aptidão_população)

    histórico_aptidão.append(aptidão_média)

    # Encontrar e imprimir a melhor solução

    melhor_indivíduo, melhor_aptidão = encontrar_melhor_indivíduo(população, aptidão_população)

    print(f'Melhor configuração: {[c["id"] for c in melhor_indivíduo]} com aptidão {melhor_aptidão}')



# Plotar gráfico da evolução dos resultados

plt.figure(figsize=(10, 6))

plt.plot(range(número_de_gerações), histórico_aptidão, marker='o', linestyle='-', color='b')

plt.title('Evolução da Aptidão Média ao Longo das Gerações')

plt.xlabel('Geração')

plt.ylabel('Aptidão Média')

plt.grid(True)

plt.show()
```
